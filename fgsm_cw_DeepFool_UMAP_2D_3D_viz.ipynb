```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.colors as mcolors
from mpl_toolkits.mplot3d import Axes3D
```

## 2D data visualization of FGSM and Carlini-Wagner attacks

```python
df = pd.read_csv("Umap_data_2D_fgsm0.01_updatedf0.csv")

# Define color mapping: 0  Green, 1  Red
colors = {0: "green", 1: "red"}
df["color"] = df["attack"].map(colors)  # Apply color mapping

cmap = mcolors.ListedColormap(["darkgreen", "red"])
bounds = [0, 0.5, 1]  # Define boundaries between colors
norm = mcolors.BoundaryNorm(bounds, cmap.N)

# Create scatter plot
plt.figure(figsize=(8, 5))
scatter = plt.scatter(df["umap0"], df["umap1"], c=df["attack"], cmap=cmap, norm=norm, edgecolors="k", alpha=0.6, s=5)

# Add colorbar with o green(legitimate, 0) and red(attacked,1)
cbar = plt.colorbar(scatter, ticks=[0, 1])
cbar.set_ticks([0, 1])
cbar.set_ticklabels(["Legitimate", "Attack"])

# Labels and title
plt.xlabel("umap0")
plt.ylabel("umap1")
plt.title("2D UMAP FGSM")
plt.savefig("XAI 2Dplot UMAP FGSM0.01.png", dpi=300, bbox_inches="tight")  # Save as PNG
plt.show()
```

## 3D data visualization of FGSM and Carlini-Wagner

**i) FGSM**

```python
# Load dataset 
df = pd.read_csv('Umap_data_3D_fgsm0.01_updatedf0.csv')
colors = {0: "green", 1: "red"}
df["color"] = df["attack"].map(colors)

cmap = mcolors.ListedColormap(["green", "red"])
bounds = [0, 0.5, 1] 
norm = mcolors.BoundaryNorm(bounds, cmap.N)

# Create a 3D scatter plot
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection="3d")
scatter = ax.scatter(df["umap0"], df["umap1"], df["umap2"], c=df["attack"], cmap=cmap, norm=norm, edgecolors="k", alpha=0.7, s=6)

# Add colorbar
cbar = fig.colorbar(scatter, ax=ax, ticks=[0, 1], shrink=0.5, aspect=10)
cbar.set_ticks([0, 1])
cbar.set_ticklabels(["Legitimate", "Attack"]) 

# Labels and title
ax.set_xlabel("umap0")
ax.set_ylabel("umap1")
ax.set_zlabel("umap2")
ax.set_title("3D UMAP FGSM")
#plt.savefig("XAI 3Dplot UMAP FGSM0.01.png", dpi=300, bbox_inches="tight")  # Save as PNG
plt.show()
```

**ii) Carlini-Wagner**
